```python
import matplotlib.pyplot as plt
```

```python
num1 = [1, 2, 3, 6]   
num2 = [2, 4, 6,7 ]   

plt.plot(num1, num2 )     
plt.show()
```

```python
#for pointing with dot as marker
num1 = [1, 2, 3, 6]   
num2 = [2, 4, 6,7 ]   

plt.plot(num1, num2, marker= 'o' )     
plt.show()
```

```python
# for pointing or marking the point with star 

num1 = [1, 2, 3, 6]   
num2 = [2, 4, 6,7 ]   

plt.plot(num1, num2, marker= '*' )     
plt.show()
```

```python
#change the color of the given points we use markerfacecolor
```

```python
num1 = [1, 2, 3, 6]   
num2 = [2, 4, 6,7 ]   

plt.plot(num1, num2, marker= 'o', markersize = '10', markerfacecolor = 'red' )     
plt.show()
```

```python
num1 = [1, 2, 3, 6]   
num2 = [2, 4, 6,7 ]   

plt.plot(num1, marker = 'o' ) 
plt.plot(num2, marker = '*')
plt.show()
```

```python
#labling of x-axis and y-axis
num1 = [1, 2, 3, 6]   
num2 = [2, 4, 6,7 ]   
plt.plot(num1, marker = 'o' ) 
plt.plot(num2, marker = '*', linestyle = 'dotted')
plt.xlabel("I am x axis", color ='red')
plt.ylabel("I am y axis", color ='red')
plt.title("demo graph" , color ='red')
plt.show()
```
