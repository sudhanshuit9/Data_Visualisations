```python
 #Line Graphn 

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = [1,2,3,4,5,6,7]
var_1 = [3,2,4,1,6,7,5]

x_1 = pd.DataFrame({"var":var , "var_1":var_1})
sns.lineplot (x="var", y ="var_1", data = x_1)
plt.show()
```

```python
plt.plot(var, var_1)
plt.show()
```

```python
sns.set_style("darkgrid")
```

```python
# Load built-in Seaborn dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Scatter plot of total_bill vs tip, with gender as color
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex", style="time")
plt.title("Scatter Plot of Bill vs Tip")
plt.show()
```

```python
# Histogram of total_bill
sns.histplot(data=tips, x="total_bill", bins=20, kde=True)
plt.title("Histogram of Total Bill")
plt.show()
```

```python
# Boxplot of tips by day
sns.boxplot(x="day", y="tip", data=tips, palette="Set2")
plt.title("Boxplot of Tips by Day")
plt.show()
```

```python
# Correlation heatmap
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pairplot of multiple variables
sns.pairplot(tips, hue="sex")
plt.show()
```
